```python
# ファイルをバイナリモードで読み込み、エンコーディングを検出
def get_encoding(file_path):
    with open(file_path, 'rb') as file:
        raw_data = file.read()
        result = chardet.detect(raw_data)
        encoding = result['encoding']
        return encoding
```

## 今日の株価をdf にして、dbを更新する

```python
# file_path から code が同じファイルを見つける
def get_db_file(db_files: list[str] ,code: str) -> pd.DataFrame:
    for file_path in db_files:
        # ファイル名を取得し、拡張子を除去
        file_name_without_ext = os.path.splitext(os.path.basename(file_path))[0]
        if file_name_without_ext == code:
            df = pd.read_csv(file_path, encoding=get_encoding(file_path))
            return df
    return None
```

```python
for daily_file in daily_files:
    daily_df = pd.read_csv(daily_file, encoding=get_encoding(daily_file))
    for index, row in daily_df.iterrows():
        # 各行のデータにアクセスする
        code = row['銘柄コード']
        row['現在日付'] = pd.to_datetime(row['現在日付'], format='%Y/%m/%d')
        db_file = get_db_file(db_files, code)
        if db_file is None or db_file.empty:
            continue # db がありません。
        #set_daily_quotes(row, db_file)
        break
    break
```
